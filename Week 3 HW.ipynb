```python
from bs4 import BeautifulSoup
import requests
import folium
from pandas.io.json import json_normalize
import numpy as np
import pandas as pd
wiki = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(wiki.content, 'html.parser')
post = soup.find('table')
post1 = pd.read_html(str(post))
```

```python
can_post = pd.DataFrame(post1[0])
can_post1 = can_post.iloc[1:289,:]
can_post2 = can_post1[can_post1[1] != "Not assigned"]
can_post3 = can_post2[can_post2[2] != "Not assigned"]
can_post3.columns = ["Postcode", "Borough", "Neighbourhood"]
can_post4 = can_post3.reset_index()
can_post5 = can_post4.iloc[:,1:4]
can_post5
```

```python
can_post5['latitude'] = np.zeros(can_post5.shape[0])
can_post5['longitude'] = np.zeros(can_post5.shape[0])
```

```python
import geocoder
# initialize your variable to None
for i in range(10):
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.google('{}, Toronto, Ontario'.format(can_post5.iloc[i, 0]))
      lat_lng_coords = g.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    print(latitude, longitude)
```
